# Image Processing Toolkit

Example notebook that showcases various functionality within the `image_processing_toolkit`


```python
# imports
from pathlib import Path

from image_processing_toolkit import (
    analysis,
    io,
    feature_extraction,
    filtering,
    transformations,
)

# Holds our test image
test_image_path = Path("../data/lena.png").resolve()
```

## IO Module


```python
# Load image
image = io.load_image(str(test_image_path))

# Display image
io.display_image(image)
```

## Transformations


```python
# Rotate image 30 degrees
image_rotated = transformations.rotate_image(image, 30)
io.display_image(image_rotated)
```

```python
# Resize image to 64x64
image_resized = transformations.resize_image(image, width=64, height=64)
io.display_image(image_resized)
```

## Feature Extraction


```python
# Extract edges
image_edges = feature_extraction.extract_edges(image)
io.display_image(image_edges)
```

```python
# Detect corners
corners, marked_image = feature_extraction.detect_corners(image, threshold=0.05)
print("Detected Corners:")
io.display_image(corners)
print("Display Strong Corners:")
io.display_image(marked_image)
```

## Filtering


```python
# Apply gaussian noise
print("Image with Gaussian Noise:")
image_gaussian_noise = transformations.apply_gaussian_noise(image, std=2.5)
io.display_image(image_gaussian_noise)

# Apply gaussian blurring to remove gaussian noise
print("Applying Gaussian Blur Filter to Remove Gaussian Noise:")
image_gaussian_blur = filtering.apply_gaussian_blur(
    image_gaussian_noise, kernel_size=(7, 7), sigmaX=10, sigmaY=10
)
io.display_image(image_gaussian_blur)
```

```python
# Apply salt and pepper noise
print("Image with Salt and Pepper Noise:")
image_snp_noise = transformations.apply_salt_and_pepper_noise(
    image, salt_prob=0.1, pepper_prob=0.1
)
io.display_image(image_snp_noise)

# Apply median blurring to remove salt and pepper noise
print("Applying Median Blur Filter to Remove Salt and Pepper Noise:")
image_median_blur = filtering.apply_median_blur(image_snp_noise, kernel_size=5)
io.display_image(image_median_blur)
```

```python
# Apply bilateral filter to remove gaussian noise
print("Applying Bilateral Filter to Remove Gaussian Noise:")
image_bilateral_filter = filtering.apply_bilateral_filter(
    image_gaussian_noise, d=9, sigma_color=375, sigma_space=10
)
io.display_image(image_bilateral_filter)
```

## Analysis


```python
# Plot histogram of image
io.display_image(image)
histogram = analysis.plot_histogram(image)
```
